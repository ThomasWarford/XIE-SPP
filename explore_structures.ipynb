# Generating structures

```python
symbols_list = ['MoS2', 'CSi', 'GaN']

structure_list = []
for symbols in symbols_list:
    structure_list += atomic_structure_generator(
        symbols=symbols,
        cspd_file='AtomicStructureGenerator/CSPD.db',
    )
```

```python
cspd_output = structure_list
cspd_output = pd.DataFrame([[
    cspd_output[i], cspd_output[i].oid, cspd_output[i].sgn, str(cspd_output[i].symbols)
] for i in range(len(cspd_output))],
                        columns=['atom', 'oid', 'sgn', 'formula'])
cspd_output['pretty_formula'] = [Formula(i).reduce()[0].format('metal') for i in cspd_output['formula']]
print(cspd_output['pretty_formula'].value_counts())
df = cspd_output
df
```

# Predicting structures

## By CNN

```python
generator = image_generator.ImageGeneratorDataFrame(df, target_col='atom')

cnn_dir = 

from tensorflow.keras.models import load_model

trained_model = load_model(cnn_dir + '/model.h5')
trained_model.load_weights(cnn_dir + '/weights0006.h5')
```

```python
yp = trained_model.predict(
    generator, 
    steps=len(generator),
    verbose=1,
    use_multiprocessing=True,
    workers=32,
    max_queue_size=10,
                             )
df['yp'] = yp.flatten()
df['ypl'] = np.sign(np.sign(yp.flatten() - 0.5) + .5)
```

## By CAE + MLP Classifier

```python
from tensorflow.keras.models import load_model

encoder = load_model(config.cae_mlp_model_dir + '/encoder.h5')
encoder.load_weights(config.cae_mlp_model_dir + '/encoder_weights.h5')

generator = image_generator.ImageGeneratorDataFrame(df, target_col='atom')
lsr = encoder.predict(
    generator, 
    steps=len(generator),
    verbose=1,
    use_multiprocessing=True,
    workers=32,
    max_queue_size=10,
                             )
```

```python
clf = load_var(config.cae_mlp_model_clf_dir + '/classifier_class.pkl')
yp = clf.predict_proba(lsr)
df['yp'] = yp.flatten()
df['ypl'] = np.sign(np.sign(yp.flatten() - 0.5) + .5)
```

## Preparing the results

```python
df['yp1'] = np.round(df['yp'], decimals=1)

print('Positive prediction -> {}/{} ({:.2f}%)'.format(np.count_nonzero(df['ypl']>0), len(df),
                                                      np.count_nonzero(df['ypl']>0)/len(df)*100))
print(df['yp1'].value_counts())

df.to_csv(result_dir + '/df.csv')
df.to_pickle(result_dir + '/df.pkl')

df.sort_values('yp')
```

# Selecting structures

```python
selected_structures = {}
formula = symbols_list[0]

for formula in symbols_list:
    df = df.sort_index()
    df = df.sample(frac=1, random_state=0)
    
    print('Formula: ', formula)
    dfs = df[df['pretty_formula'] == formula].sort_values('yp1')
#     dfs = dfs.drop_duplicates('sgn')
#     assert ~dfs['sgn'].duplicated().any()
    
    print('Positive prediction -> {}/{} ({:.2f}%)'.format(np.count_nonzero(dfs['ypl']>0), len(dfs),
                                                      np.count_nonzero(dfs['ypl']>0)/len(dfs)*100))
    print(dfs['yp1'].value_counts())
    
    dfs['path'] = [result_dir + f'/cif/{formula}/{formula}_{i}.cif' for i in dfs['sgn']]
    _, ind = np.unique(dfs['yp1'], return_index=True)
    print('Selected structures for displaying ({} structures)'.format(len(ind)))
    display(dfs.iloc[ind, :])
    selected_structures[formula] = dfs.iloc[ind, :]
save_var(selected_structures, result_dir + '/selected_structures.pkl')
```
