## Most well studied compositions

```python
compositions_plus_cod = pd.read_csv('mat2vec-master/formula no filter.txt', sep='\t', index_col=0)
df = compositions_plus_cod
df['composition'] = [i.strip() for i in df['formula']]
df['rank'] = np.array(range(len(df)), dtype=int) + 1
df['n_repetition'] = df['literature']
df.to_csv('compositions_in_literature.csv', index=False)
```

```python
sns.barplot(data=df.head(20), x='composition', y='n_repetition')
plt.xticks(rotation=90)
plt.title('Top 20 compositions')
```

```python
sns.lineplot(data=df.head(200), x='rank', y='n_repetition')
# plt.xticks(rotation=90)
plt.title('Top 200 compositions')
```

```python
x1 = 50
x2 = 1000

sns.lineplot(data=df[(df['rank'] >= x1) & (df['rank'] < x2)], x='rank', y='n_repetition')
plt.xticks(np.arange(x1, x2, 50))
plt.xticks(rotation=90)
plt.title(f'Top {x1}-{x2} compositions')
```

```python
x1 = 500
x2 = 5000

sns.lineplot(data=df[(df['rank'] >= x1) & (df['rank'] < x2)], x='rank', y='n_repetition')
plt.xticks(np.arange(x1, x2, 500))
plt.xticks(rotation=90)
plt.title(f'Top {x1}-{x2} compositions')
```

```python
sns.lineplot(data=df[df['n_repetition'] >= 1000], x='rank', y='n_repetition')
# plt.xticks(np.arange(x1, x2, 500))
plt.xticks(rotation=90)
plt.title(f'Top {x1}-{x2} compositions')
```

## Logarithmic plots

```python
x1 = 0
x2 = 5000

sns.lineplot(data=df[(df['rank'] >= x1) & (df['rank'] < x2)], x='rank', y='n_repetition')
plt.xticks(np.arange(x1, x2, 500))
plt.xticks(rotation=90)
plt.yscale('log', base=10)
plt.title(f'Top {x1}-{x2} compositions')
```
